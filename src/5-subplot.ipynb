### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.


import matplotlib.pyplot as plt
import numpy as np

# Data for plots
x = np.linspace(0, 10, 100)
y = np.sin(x)
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Line plot
axes[0, 0].plot(x, y, color='blue')
axes[0, 0].set_title("Line Chart")
axes[0, 0].set_xlabel("X Axis")
axes[0, 0].set_ylabel("Y Axis")

# Bar plot
axes[0, 1].bar(categories, values, color=['blue', 'green', 'red', 'purple'])
axes[0, 1].set_title("Bar Chart")
axes[0, 1].set_xlabel("Categories")
axes[0, 1].set_ylabel("Values")

# Scatter plot
axes[1, 0].scatter(scatter_x, scatter_y, color='red')
axes[1, 0].set_title("Scatter Plot")
axes[1, 0].set_xlabel("X Values")
axes[1, 0].set_ylabel("Y Values")

# Pie chart
axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'green', 'red', 'orange'], startangle=140)
axes[1, 1].set_title("Pie Chart")

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()


```
