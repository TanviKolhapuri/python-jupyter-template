### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.

import matplotlib.pyplot as plt

# Data
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

# Error handling: Check if data is not empty
if not data:
    raise ValueError("Data for the histogram is empty. Please provide valid input.")

# Create histogram
plt.hist(data, bins=5, color='indigo', edgecolor='black')

# Customize plot
plt.title('Histogram of Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show plot
plt.show()
```
